Probar que el sistema dx/dt=y+x(a-x^2-y^2), dy/dt=-x+y(1-x^2-y^2) con 0<a<1 tiene un ciclo limite y determinar su estabilidad.

```python
from sympy import *
from sympy.abc import x,y,t,r,a,theta
```

Sistema en coordenadas polares: 
dr/dt=r(acos(theta)^2+sin(theta)^2)-r^2
dtheta/dt=-1+cos(theta)sin(theta)(1-a)

```python
#puntos criticos
P=r*(a*cos(theta)**2+sin(theta)**2)-r**2 
Q=-1+(cos(theta)*sin(theta)*(1-a))
#establecer P(x,y)=0 y Q(x,y)=0
Peqn=Eq(P,0)
Qeqn=Eq(Q,0)
print(solve((Peqn,Qeqn),r,theta))
```

```python
#matriz Jacobiana
J11=diff(P,r)
J12=diff(P,theta)
J21=diff(Q,r)
J22=diff(Q,theta)
J=Matrix([[J11,J12],[J21,J22]])
pprint(J)
```

```python
#J en el punto critico
Jc1=J.subs([(r,0),(theta,0)])
print(Jc1.eigenvals())
pprint(Jc1.eigenvects())
Jc2=J.subs([(r,0),(theta,-atan(a))])
print(Jc2.eigenvals())
pprint(Jc2.eigenvects())
```

El punto critico (0,theta) tiene eigenvalores reales posivios distintos, por lo que es un nodo inestable. El punto critico (0,-atan(a)) tiene eigenvalores iguales de multiplicidad 2, con un solo eigenvector linealmente independiente por lo que es un nodo degenerado.
