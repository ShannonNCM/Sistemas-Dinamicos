Ciclo limite para modelo Fitzhugh-Nagumo

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
theta=0.14
omega=0.112
gamma=2.54
epsilon=0.01
xmin=-0.5
xmax=1.5
ymin=0
ymax=0.3
```

```python
def dx_dt(x,t):
    return [ -x[0]*(x[0]-theta)*(x[0]-1)-x[1]+omega , epsilon*(x[0]-gamma*x[1]) ]
```

```python
#trayectorias en tiempo hacia adelante
xs=odeint(dx_dt,[0.5,0.09],np.linspace(0,100,1000))
plt.plot(xs[:,0],xs[:,1],"r-")
#etiquetas de ejes y tama;o de letra
plt.xlabel('u',fontsize=15)
plt.ylabel('v',fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)
#nullclines
x=np.arange(xmin,xmax,0.01)
plt.plot(x,x/gamma,'b--',x,-x*(x-theta)*(x-1)+omega,'b--')
plt.show()
```
