```python
from sympy import *
from sympy.abc import S,I,F,t,a,b,c,N
```

```python
S=Function('S')
I=Function('I')
F=Function('F')
```

```python
eqsd=[Eq(S(t).diff(t),-a*S(t)+b*I(t)),Eq(I(t).diff(t),a*S(t)-b*I(t)-c*I(t)),Eq(F(t).diff(t),c*I(t))]
```

```python
sol=dsolve(eqsd,[S(t),I(t),F(t)])
print(sol)
```

```python
sol=dsolve(eqsd,[S(t),I(t),F(t)])
pprint(sol)
```

```python
sol=dsolve(eqsd,[S(t),I(t),F(t)],ics={S(0):N,I(0):0,F(0):0})
pprint(sol)
```
