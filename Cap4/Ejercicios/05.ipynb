```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pylab as pl
```

a) Retrato de fase del sistema dx/dt=x(4-y-x) dy/dt=y(3x-1-y) para x,y>=0  dados los punto criticos en (0,0) (0,4) y (5/4,11/4)

```python
#sistema 2D no lineal
def dx_dt(x,t):
    return [ x[0]*(4-x[1]-x[0]) , x[1]*(3*x[0]-1-x[1]) ]
#trayectorias en tiempo hacia adelante
ts=np.linspace(0,10,500)
ic=np.linspace(0,5,5)
for r in ic:
    for s in ic:
        x0=[r,s]
        xs=odeint(dx_dt,x0,ts)
        plt.plot(xs[:,0],xs[:,1],"r-")
#trayectorias en tiempo hacia atras
ts=np.linspace(0,-10,500)
ic=np.linspace(0,5,5)
for r in ic:
    for s in ic:
        x0=[r,s]
        xs=odeint(dx_dt,x0,ts)
        plt.plot(xs[:,0],xs[:,1],"r-")
#etiquetas de ejes y estilo de letra
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
plt.tick_params(labelsize=10)
plt.xlim(0,5)
plt.ylim(0,5)
#campo vectorial
#X,Y=np.mgrid[0:5:20j,0:5:20j]
#u=X*(4-Y-X)
#v=Y*(3*X-1-Y)
#pl.quiver(X,Y,u,v,color='b')
plt.show()
```

b) Retrato de fase del sistema dx/dt=x(2-y-x) dy/dt=y(3-2x-y) para x,y>=0 dados los punto criticos en (0,0) (0,4) y (5/4,11/4)

```python
#sistema 2D no lineal
def dx2_dt(x,t):
    return [ x[0]*(2-x[1]-x[0]) , x[1]*(3-2*x[0]-x[1]) ]
#trayectorias en tiempo hacia adelante
ts=np.linspace(0,10,500)
ic=np.linspace(0,5,5)
for r in ic:
    for s in ic:
        x0=[r,s]
        xs=odeint(dx2_dt,x0,ts)
        plt.plot(xs[:,0],xs[:,1],"r-")
#trayectorias en tiempo hacia atras
ts=np.linspace(0,-10,500)
ic=np.linspace(0,5,5)
for r in ic:
    for s in ic:
        x0=[r,s]
        xs=odeint(dx2_dt,x0,ts)
        plt.plot(xs[:,0],xs[:,1],"r-")
#etiquetas de ejes y estilo de letra
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
plt.tick_params(labelsize=10)
plt.xlim(0,5)
plt.ylim(0,5)
#campo vectorial
#X,Y=np.mgrid[0:5:20j,0:5:20j]
#u=X*(2-Y-X)
#v=Y*(3-2*X-Y)
#pl.quiver(X,Y,u,v,color='b')
plt.show()
```

El modelo del inciso a) representa la interaccion depredador-presa, donde llegan al punto estable en el punto critico (5/4,11/4).
El modelo del inciso b) representa la interaccion de especies que compiten, en la cual una de ellas prevalece sobre la ot.
