```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pylab as pl
```

```python
#sistema 2D no lineal
def dx_dt(x,t):
    return [ 2*x[0]-x[0]**2-x[0]*x[1] , 3*x[1]-x[1]**2-2*x[0]*x[1] ]
#trayectorias en tiempo hacia adelante
ts=np.linspace(0,10,500)
ic=np.linspace(0,5,5)
for r in ic:
    for s in ic:
        x0=[r,s]
        xs=odeint(dx_dt,x0,ts)
        plt.plot(xs[:,0],xs[:,1],"r-")
#trayectorias en tiempo hacia atras
ts=np.linspace(0,-10,500)
ic=np.linspace(0,5,5)
for r in ic:
    for s in ic:
        x0=[r,s]
        xs=odeint(dx_dt,x0,ts)
        plt.plot(xs[:,0],xs[:,1],"r-")
#etiquetas de ejes y estilo de letra
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
plt.tick_params(labelsize=10)
plt.xlim(0,5)
plt.ylim(0,5)
#campo vectorial
X,Y=np.mgrid[0:5:20j,0:5:20j]
u=2*X-X**2-X*Y
v=3*Y-Y**2-2*X*Y
pl.quiver(X,Y,u,v,color='b')
plt.show()
```

El sistema es de especies que compiten, del retrato de fase se observa que una de las especies se extingue.
